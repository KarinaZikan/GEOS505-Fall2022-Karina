```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/gapminder_gdp_africa.csv', index_col='country')
df.head()
df = df.head()
```

```python
years = df.columns.str.strip('gdpPercap_')
years
```

```python
df.columns = years.astype(int)
df.head()
```

```python
df.loc['Algeria'].plot()
```

```python
df.T.plot()
plt.ylabel('gdp per capita')
```

```python
plt.style.use('ggplot')
df.T.plot(kind='bar')
plt.ylabel('GDP per capita')
```

```python
years = df.columns
gdp_algeria = df.loc['Algeria']

plt.plot(years, gdp_algeria, 'g--')
```

```python
# Select two countries' worth of data.
gdp_algeria = df.loc['Algeria']
gdp_angola = df.loc['Angola']

# Plot with differently-colored markers.
plt.plot(years, gdp_algeria, 'b-', label='Algeria')
plt.plot(years, gdp_angola, 'g-', label='Angola')

# Create legend.
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')
```
